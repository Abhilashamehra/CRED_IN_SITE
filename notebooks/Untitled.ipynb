```python
# Get basic data types, null counts, and memory usage
df.info()

# Basic statistics for numeric columns
df.describe()
```

```python
# Total missing count
missing_counts = df.isnull().sum()

# Percent missing
missing_percent = (missing_counts / len(df)) * 100

# Summary DataFrame
missing_summary = pd.DataFrame({
    'Missing Values': missing_counts,
    'Percentage(%)': missing_percent
})

# Filter non-zero missing and sort
missing_summary = missing_summary[missing_summary['Missing Values'] > 0]
missing_summary.sort_values(by='Percentage(%)', ascending=False, inplace=True)

# Top 20
missing_summary.head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 20 missing value columns
top_missing = missing_summary.head(20)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_missing['Percentage(%)'], y=top_missing.index, palette="Reds_r")
plt.title("Top 20 Columns with Most Missing Values")
plt.xlabel("Percentage Missing")
plt.ylabel("Column")
plt.tight_layout()
plt.show()
```
