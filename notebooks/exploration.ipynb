```python
import pandas as pd

# Loading training data

df = pd.read_csv("/Users/abby/Projects/credinsite/data/raw/application_train.csv")

# Viewing shape 
print ("Shape:", df.shape)
df.head()
```

```python
#Understanding datatypes and stats
df.info()
df.describe()
```

```python
# Checking total missing values per column
missing_counts = df.isnull().sum()

# Calculating % of missing values
missing_percent = (missing_counts/len(df))*100

#Combining both into a summary dataframe
missing_summary = pd.DataFrame(
   {
       'Missing Values': missing_counts,
       'Percentage(%)': missing_percent
   } 
)

#Filtering columns with missing values
missing_summary = missing_summary[missing_summary['Missing Values']>0]

#Sort by percentage descending
missing_summary.sort_values(by='Percentage(%)', ascending=False, inplace=True)

#Display the top 20
missing_summary.head(20)
```

```python
# drop columns with more than 50% missing values
```

```python
cols_to_drop = missing_summary[missing_summary['Percentage(%)']>50].index.tolist()
df_cleaned = df.drop(columns = cols_to_drop)
```

```python
missing_cleaned = df_cleaned.isnull().sum()
missing_cleaned = missing_cleaned[missing_cleaned >0].sort_values(ascending=False)
missing_cleaned
```

```python
# Identify categorial columns which still have missing values 
cat_cols_missing = df_cleaned[missing_cleaned.index].select_dtypes(include='object').columns.tolist()

#Numerical columns with missing values
num_cols_missing = [col for col in missing_cleaned.index if col not in cat_cols_missing]

print("Categorial Columns:", cat_cols_missing)
print("Numerical Columns:", num_cols_missing)
```

```python
for col in cat_cols_missing:
    mode_val= df_cleaned[col].mode()[0]
    df_cleaned[col] = df_cleaned[col].fillna(mode_val)
```

```python
for col in num_cols_missing:
    median_val = df_cleaned[col].median()
    df_cleaned[col] = df_cleaned[col].fillna(median_val)
```
